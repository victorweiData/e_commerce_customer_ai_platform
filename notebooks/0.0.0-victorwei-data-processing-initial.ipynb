# Customer Segmentation

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
class CustomerSegmentation:
    def __init__(self, df_orders, df_order_items, df_payments, df_reviews, df_customers, df_products, df_categories):
        self.df_orders = df_orders
        self.df_order_items = df_order_items
        self.df_payments = df_payments
        self.df_reviews = df_reviews
        self.df_customers = df_customers
        self.df_products = df_products
        self.df_categories = df_categories
        
        # Convert date columns
        self.df_orders['order_purchase_timestamp'] = pd.to_datetime(self.df_orders['order_purchase_timestamp'])
        
        # Set analysis date (use max date + 1 day for recency calculation)
        self.analysis_date = self.df_orders['order_purchase_timestamp'].max() + timedelta(days=1)
        
    def create_customer_master_table(self):
        """Create a comprehensive customer master table with all metrics"""
        print("Creating customer master table...")
        
        # Start with orders data
        customer_orders = self.df_orders.groupby('customer_id').agg({
            'order_id': 'count',
            'order_purchase_timestamp': ['min', 'max']
        }).round(2)
        
        # Flatten column names
        customer_orders.columns = ['frequency', 'first_order_date', 'last_order_date']
        
        # Calculate recency (days since last order)
        customer_orders['recency'] = (self.analysis_date - customer_orders['last_order_date']).dt.days
        
        # Calculate customer tenure (days between first and last order)
        customer_orders['tenure_days'] = (customer_orders['last_order_date'] - customer_orders['first_order_date']).dt.days
        customer_orders['tenure_days'] = customer_orders['tenure_days'].fillna(0)  # For single-order customers
        
        # Merge with order items to get monetary metrics
        order_monetary = self.df_order_items.groupby('order_id').agg({
            'price': 'sum',
            'freight_value': 'sum'
        }).reset_index()
        order_monetary['total_value'] = order_monetary['price'] + order_monetary['freight_value']
        
        # Merge with orders to get customer_id
        order_customer_value = self.df_orders[['order_id', 'customer_id']].merge(
            order_monetary, on='order_id'
        )
        
        # Calculate monetary metrics per customer
        customer_monetary = order_customer_value.groupby('customer_id').agg({
            'price': ['sum', 'mean'],
            'freight_value': ['sum', 'mean'],
            'total_value': ['sum', 'mean']
        }).round(2)
        
        # Flatten column names
        customer_monetary.columns = [
            'total_price', 'avg_order_value', 'total_freight', 'avg_freight',
            'total_spent', 'avg_total_value'
        ]
        
        # Merge RFM data
        customer_master = customer_orders.merge(customer_monetary, left_index=True, right_index=True)
        
        # Add payment behavior
        payment_behavior = self.get_payment_behavior()
        customer_master = customer_master.merge(payment_behavior, left_index=True, right_index=True, how='left')
        
        # Add review behavior
        review_behavior = self.get_review_behavior()
        customer_master = customer_master.merge(review_behavior, left_index=True, right_index=True, how='left')
        
        # Add product category behavior
        category_behavior = self.get_category_behavior()
        customer_master = customer_master.merge(category_behavior, left_index=True, right_index=True, how='left')
        
        # Add geographic info
        geo_info = self.get_geographic_info()
        customer_master = customer_master.merge(geo_info, left_index=True, right_index=True, how='left')
        
        # Fill NaN values
        customer_master = customer_master.fillna(0)
        
        self.customer_master = customer_master
        print(f"Customer master table created with {len(customer_master)} customers and {len(customer_master.columns)} features")
        return customer_master
    
    def get_payment_behavior(self):
        """Get payment method preferences and behavior"""
        # Merge payments with orders to get customer_id
        payment_customer = self.df_orders[['order_id', 'customer_id']].merge(
            self.df_payments, on='order_id'
        )
        
        payment_behavior = payment_customer.groupby('customer_id').agg({
            'payment_type': lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else 'credit_card',
            'payment_installments': 'mean',
            'payment_value': 'sum'
        }).round(2)
        
        payment_behavior.columns = ['preferred_payment_type', 'avg_installments', 'total_payment_value']
        
        # Add payment diversity (number of different payment types used)
        payment_diversity = payment_customer.groupby('customer_id')['payment_type'].nunique()
        payment_behavior['payment_diversity'] = payment_diversity
        
        return payment_behavior
    
    def get_review_behavior(self):
        """Get review giving behavior and satisfaction scores"""
        # Merge reviews with orders to get customer_id
        review_customer = self.df_orders[['order_id', 'customer_id']].merge(
            self.df_reviews, on='order_id', how='left'
        )
        
        review_behavior = review_customer.groupby('customer_id').agg({
            'review_score': ['mean', 'count'],
            'review_comment_message': lambda x: x.notna().sum()  # Count of reviews with comments
        }).round(2)
        
        # Flatten column names
        review_behavior.columns = ['avg_review_score', 'review_count', 'reviews_with_comments']
        
        # Calculate review rate (reviews per order)
        customer_order_counts = self.df_orders.groupby('customer_id').size()
        review_behavior['review_rate'] = (review_behavior['review_count'] / customer_order_counts).fillna(0).round(2)
        
        return review_behavior
    
    def get_category_behavior(self):
        """Get product category preferences and diversity"""
        # Merge products with category translation
        products_with_categories = self.df_products.merge(
            self.df_categories, on='product_category_name', how='left'
        )
        
        # Merge order items with products and orders
        items_with_categories = self.df_order_items.merge(
            products_with_categories[['product_id', 'product_category_name_english']], 
            on='product_id', how='left'
        ).merge(
            self.df_orders[['order_id', 'customer_id']], on='order_id'
        )
        
        # Calculate category metrics per customer
        category_behavior = items_with_categories.groupby('customer_id').agg({
            'product_category_name_english': ['nunique', lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else 'other']
        }).round(2)
        
        # Flatten column names
        category_behavior.columns = ['category_diversity', 'preferred_category']
        
        return category_behavior
    
    def get_geographic_info(self):
        """Get customer geographic information"""
        geo_info = self.df_customers.set_index('customer_id')[['customer_city', 'customer_state']]
        return geo_info
    
    def calculate_rfm_scores(self):
        """Calculate RFM scores using quintile-based scoring"""
        print("Calculating RFM scores...")
        
        if not hasattr(self, 'customer_master'):
            self.create_customer_master_table()
        
        # Calculate quintile-based scores (1-5, where 5 is best)
        self.customer_master['R_score'] = pd.qcut(
            self.customer_master['recency'], q=5, labels=[5,4,3,2,1]
        ).astype(int)
        
        self.customer_master['F_score'] = pd.qcut(
            self.customer_master['frequency'].rank(method='first'), q=5, labels=[1,2,3,4,5]
        ).astype(int)
        
        self.customer_master['M_score'] = pd.qcut(
            self.customer_master['total_spent'].rank(method='first'), q=5, labels=[1,2,3,4,5]
        ).astype(int)
        
        # Combined RFM score
        self.customer_master['RFM_score'] = (
            self.customer_master['R_score'].astype(str) + 
            self.customer_master['F_score'].astype(str) + 
            self.customer_master['M_score'].astype(str)
        )
        
        print("RFM scores calculated successfully")
        return self.customer_master[['R_score', 'F_score', 'M_score', 'RFM_score']]
    
    def create_rfm_segments(self):
        """Create RFM-based customer segments"""
        print("Creating RFM segments...")
        
        if 'RFM_score' not in self.customer_master.columns:
            self.calculate_rfm_scores()
        
        def rfm_segment(row):
            if row['RFM_score'] in ['555', '554', '544', '545', '454', '455', '445']:
                return 'Champions'
            elif row['RFM_score'] in ['543', '444', '435', '355', '354', '345', '344', '335']:
                return 'Loyal Customers'
            elif row['RFM_score'] in ['512', '511', '422', '421', '412', '411', '311']:
                return 'Potential Loyalists'
            elif row['RFM_score'] in ['553', '551', '552', '541', '542', '533', '532', '531', '452', '451']:
                return 'New Customers'
            elif row['RFM_score'] in ['155', '154', '144', '214', '215', '115', '114']:
                return 'Promising'
            elif row['RFM_score'] in ['155', '254', '245', '253', '252', '243', '244', '235', '234', '225', '224', '153', '152']:
                return 'Need Attention'
            elif row['RFM_score'] in ['331', '321', '231', '241', '251']:
                return 'About to Sleep'
            elif row['RFM_score'] in ['155', '144', '214', '215', '115', '114', '113']:
                return 'At Risk'
            elif row['RFM_score'] in ['125', '124', '123', '122', '132', '131', '121', '212', '211']:
                return 'Cannot Lose Them'
            elif row['RFM_score'] in ['432', '423', '353', '334', '343', '323', '313']:
                return 'Hibernating'
            else:
                return 'Lost'
        
        self.customer_master['RFM_segment'] = self.customer_master.apply(rfm_segment, axis=1)
        
        segment_counts = self.customer_master['RFM_segment'].value_counts()
        print("RFM Segment Distribution:")
        for segment, count in segment_counts.items():
            print(f"  {segment}: {count:,} ({count/len(self.customer_master)*100:.1f}%)")
        
        return self.customer_master['RFM_segment']
    
    def create_value_segments(self):
        """Create value-based segments"""
        print("Creating value-based segments...")
        
        if not hasattr(self, 'customer_master'):
            self.create_customer_master_table()
        
        # Define value segments based on total spent
        value_percentiles = self.customer_master['total_spent'].quantile([0.25, 0.50, 0.75, 0.90])
        
        def value_segment(total_spent):
            if total_spent >= value_percentiles[0.90]:
                return 'VIP'
            elif total_spent >= value_percentiles[0.75]:
                return 'High Value'
            elif total_spent >= value_percentiles[0.50]:
                return 'Medium Value'
            elif total_spent >= value_percentiles[0.25]:
                return 'Low Value'
            else:
                return 'Entry Level'
        
        self.customer_master['value_segment'] = self.customer_master['total_spent'].apply(value_segment)
        
        segment_counts = self.customer_master['value_segment'].value_counts()
        print("Value Segment Distribution:")
        for segment, count in segment_counts.items():
            print(f"  {segment}: {count:,} ({count/len(self.customer_master)*100:.1f}%)")
        
        return self.customer_master['value_segment']
    
    def create_behavioral_segments(self):
        """Create behavioral segments using clustering"""
        print("Creating behavioral segments...")
        
        if not hasattr(self, 'customer_master'):
            self.create_customer_master_table()
        
        # Select behavioral features
        behavioral_features = [
            'frequency', 'avg_order_value', 'tenure_days', 'avg_installments',
            'payment_diversity', 'avg_review_score', 'review_rate', 'category_diversity'
        ]
        
        # Prepare data for clustering
        behavioral_data = self.customer_master[behavioral_features].fillna(0)
        
        # Standardize features
        scaler = StandardScaler()
        behavioral_scaled = scaler.fit_transform(behavioral_data)
        
        # Determine optimal number of clusters using elbow method
        sse = []
        k_range = range(2, 11)
        for k in k_range:
            kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
            kmeans.fit(behavioral_scaled)
            sse.append(kmeans.inertia_)
        
        # Use 5 clusters as default (can be adjusted)
        optimal_k = 5
        kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
        behavioral_clusters = kmeans.fit_predict(behavioral_scaled)
        
        self.customer_master['behavioral_cluster'] = behavioral_clusters
        
        # Analyze clusters to give meaningful names
        cluster_profiles = self.customer_master.groupby('behavioral_cluster')[behavioral_features].mean()
        
        # Create meaningful cluster names based on characteristics
        cluster_names = {}
        for cluster in range(optimal_k):
            profile = cluster_profiles.loc[cluster]
            
            if profile['frequency'] > cluster_profiles['frequency'].mean() and profile['avg_order_value'] > cluster_profiles['avg_order_value'].mean():
                cluster_names[cluster] = 'High Frequency High Value'
            elif profile['frequency'] > cluster_profiles['frequency'].mean():
                cluster_names[cluster] = 'Frequent Buyers'
            elif profile['avg_order_value'] > cluster_profiles['avg_order_value'].mean():
                cluster_names[cluster] = 'Big Spenders'
            elif profile['category_diversity'] > cluster_profiles['category_diversity'].mean():
                cluster_names[cluster] = 'Category Explorers'
            else:
                cluster_names[cluster] = 'Occasional Shoppers'
        
        self.customer_master['behavioral_segment'] = self.customer_master['behavioral_cluster'].map(cluster_names)
        
        segment_counts = self.customer_master['behavioral_segment'].value_counts()
        print("Behavioral Segment Distribution:")
        for segment, count in segment_counts.items():
            print(f"  {segment}: {count:,} ({count/len(self.customer_master)*100:.1f}%)")
        
        return self.customer_master['behavioral_segment']
    
    def analyze_segments(self):
        """Analyze all segment types and create summary"""
        print("Analyzing all segments...")
        
        # Create all segments
        self.create_rfm_segments()
        self.create_value_segments()
        self.create_behavioral_segments()
        
        # Create comprehensive analysis
        segment_analysis = {}
        
        # RFM Segment Analysis
        rfm_analysis = self.customer_master.groupby('RFM_segment').agg({
            'recency': 'mean',
            'frequency': 'mean',
            'total_spent': 'mean',
            'avg_order_value': 'mean',
            'avg_review_score': 'mean'
        }).round(2)
        segment_analysis['RFM'] = rfm_analysis
        
        # Value Segment Analysis
        value_analysis = self.customer_master.groupby('value_segment').agg({
            'total_spent': ['mean', 'min', 'max'],
            'frequency': 'mean',
            'avg_order_value': 'mean',
            'recency': 'mean'
        }).round(2)
        segment_analysis['Value'] = value_analysis
        
        # Behavioral Segment Analysis
        behavioral_analysis = self.customer_master.groupby('behavioral_segment').agg({
            'frequency': 'mean',
            'avg_order_value': 'mean',
            'category_diversity': 'mean',
            'payment_diversity': 'mean',
            'avg_review_score': 'mean',
            'total_spent': 'mean'
        }).round(2)
        segment_analysis['Behavioral'] = behavioral_analysis
        
        self.segment_analysis = segment_analysis
        return segment_analysis
    
    def plot_segments(self):
        """Create visualizations for all segment types"""
        if not hasattr(self, 'segment_analysis'):
            self.analyze_segments()
        
        fig, axes = plt.subplots(2, 3, figsize=(20, 12))
        fig.suptitle('Customer Segmentation Analysis', fontsize=16, fontweight='bold')
        
        # RFM Segment Distribution
        rfm_counts = self.customer_master['RFM_segment'].value_counts()
        axes[0, 0].pie(rfm_counts.values, labels=rfm_counts.index, autopct='%1.1f%%', startangle=90)
        axes[0, 0].set_title('RFM Segment Distribution')
        
        # Value Segment Distribution
        value_counts = self.customer_master['value_segment'].value_counts()
        axes[0, 1].pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%', startangle=90)
        axes[0, 1].set_title('Value Segment Distribution')
        
        # Behavioral Segment Distribution
        behavioral_counts = self.customer_master['behavioral_segment'].value_counts()
        axes[0, 2].pie(behavioral_counts.values, labels=behavioral_counts.index, autopct='%1.1f%%', startangle=90)
        axes[0, 2].set_title('Behavioral Segment Distribution')
        
        # RFM Scores Distribution
        rfm_metrics = self.customer_master[['recency', 'frequency', 'total_spent']].mean()
        axes[1, 0].bar(rfm_metrics.index, rfm_metrics.values)
        axes[1, 0].set_title('Average RFM Metrics')
        axes[1, 0].set_ylabel('Values')
        
        # Segment Comparison - Total Spent
        segment_spending = self.customer_master.groupby('RFM_segment')['total_spent'].mean().sort_values(ascending=False)
        axes[1, 1].bar(range(len(segment_spending)), segment_spending.values)
        axes[1, 1].set_title('Average Spending by RFM Segment')
        axes[1, 1].set_xticks(range(len(segment_spending)))
        axes[1, 1].set_xticklabels(segment_spending.index, rotation=45, ha='right')
        axes[1, 1].set_ylabel('Average Total Spent')
        
        # Customer Lifetime Value by Segment
        clv_by_segment = self.customer_master.groupby('value_segment')['total_spent'].mean()
        axes[1, 2].bar(clv_by_segment.index, clv_by_segment.values)
        axes[1, 2].set_title('Average Spending by Value Segment')
        axes[1, 2].set_ylabel('Average Total Spent')
        axes[1, 2].tick_params(axis='x', rotation=45)
        
        plt.tight_layout()
        plt.show()
        
        return fig
    
    def get_segment_insights(self):
        """Generate actionable insights for each segment"""
        if not hasattr(self, 'segment_analysis'):
            self.analyze_segments()
        
        insights = {
            'Champions': 'Reward them. Can be early adopters for new products. Will promote your brand.',
            'Loyal Customers': 'Upsell higher value products. Ask for reviews. Engage them.',
            'Potential Loyalists': 'Offer membership or loyalty program. Recommend other products.',
            'New Customers': 'Provide on-boarding support, give them early success, start building relationship.',
            'Promising': 'Create brand awareness, offer free trials.',
            'Need Attention': 'Make limited time offers. Recommend based on past purchases. Reactivate them.',
            'About to Sleep': 'Share valuable resources, recommend popular products / renewals at discount, reconnect with them.',
            'At Risk': 'Send personalized emails to reconnect, offer renewals, provide helpful resources.',
            'Cannot Lose Them': 'Win them back via renewals or newer products, don\'t lose them to competition, talk to them.',
            'Hibernating': 'Offer other product categories, special discounts, recreate brand value.',
            'Lost': 'Revive interest with reach out campaign, ignore otherwise.'
        }
        
        segment_insights = {}
        for segment in self.customer_master['RFM_segment'].unique():
            if segment in insights:
                segment_insights[segment] = insights[segment]
        
        return segment_insights
```

```python
# Initialize the segmentation analysis
segmentation = CustomerSegmentation(
    df_orders, df_order_items, df_payments, df_reviews, 
    df_customers, df_products, df_categories
)

# Create master customer table
customer_master = segmentation.create_customer_master_table()

# Perform all segmentation analyses
segment_analysis = segmentation.analyze_segments()
```

```python
# Create visualizations
segmentation.plot_segments()

# Get actionable insights
insights = segmentation.get_segment_insights()

print("\n" + "="*50)
print("ACTIONABLE INSIGHTS BY SEGMENT")
print("="*50)
for segment, insight in insights.items():
    count = len(segmentation.customer_master[segmentation.customer_master['RFM_segment'] == segment])
    print(f"\n{segment} ({count:,} customers):")
    print(f"  Strategy: {insight}")

print(f"\nTotal customers analyzed: {len(customer_master):,}")
print(f"Total features created: {len(customer_master.columns)}")
```
